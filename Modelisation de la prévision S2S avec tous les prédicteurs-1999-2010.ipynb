# La rgression Ridge

La rgression Ridge est une technique de rgression linaire rgularise qui a t offerte pour surmonter les problmes de surajustement (overfitting) dans la modlisation de donnes. Elle est galement connue sous le nom de rgression Tikhonov en hommage au mathmaticien russe Andrei Tikhonov qui a dvelopp cette mthode.

La rgression Ridge ajoute une pnalit  la fonction de cot de la rgression linaire classique en utilisant une rgularisation L2. Cette proposition ajoute un terme qui est proportionnel au carr de la norme L2 des coefficients de la rgression. Ainsi, la fonction de cot de la rgression Ridge est :

J(w) = RSS(w) + alpha * ||w||2

O RSS(w) est la somme des quivalents des erreurs rsiduelles entre les prdictions du modle et les valeurs rvles, ||w||2 est la norme L2 des coefficients du modle, et alpha est un hyperparamtre qui contrle l'importance de la rgularisation.

En ajoutant cette pnalit L2, la rgression Ridge rduit la magnitude des coefficients de rgression et limine les variables importantes du modle, en vitant le surajustement et en rduisant ainsi la capacit de gnralisation du modle. Cependant, la rgression Ridge ne slectionne pas automatiquement les variables pertinentes, et la valeur de l'hyperparamtre alpha doit tre choisie avec soin pour optimiser la performance du modle sur les donnes.

```python
# Fonction pour calculer le coefficient de corrlation avec gestion des NaN
def correlation_coefficient(y_true, y_pred):
    if np.var(y_true) == 0 or np.var(y_pred) == 0:
        return 0  # Si la variance est nulle, la corrlation est 0
    else:
        return pearsonr(y_true, y_pred)[0]
```
