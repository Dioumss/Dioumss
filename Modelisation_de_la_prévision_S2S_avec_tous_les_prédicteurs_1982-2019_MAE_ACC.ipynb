# La rgression Ridge

La rgression Ridge est une technique de rgression linaire rgularise qui a t offerte pour surmonter les problmes de surajustement (overfitting) dans la modlisation de donnes. Elle est galement connue sous le nom de rgression Tikhonov en hommage au mathmaticien russe Andrei Tikhonov qui a dvelopp cette mthode.

La rgression Ridge ajoute une pnalit  la fonction de cot de la rgression linaire classique en utilisant une rgularisation L2. Cette proposition ajoute un terme qui est proportionnel au carr de la norme L2 des coefficients de la rgression. Ainsi, la fonction de cot de la rgression Ridge est :

J(w) = RSS(w) + alpha * ||w||2

O RSS(w) est la somme des quivalents des erreurs rsiduelles entre les prdictions du modle et les valeurs rvles, ||w||2 est la norme L2 des coefficients du modle, et alpha est un hyperparamtre qui contrle l'importance de la rgularisation.

En ajoutant cette pnalit L2, la rgression Ridge rduit la magnitude des coefficients de rgression et limine les variables importantes du modle, en vitant le surajustement et en rduisant ainsi la capacit de gnralisation du modle. Cependant, la rgression Ridge ne slectionne pas automatiquement les variables pertinentes, et la valeur de l'hyperparamtre alpha doit tre choisie avec soin pour optimiser la performance du modle sur les donnes.

```python
from sklearn.linear_model import Ridge
# Cration d'un objet Ridge Regression
Reg_R = Ridge()

# Paramtres  tester pour la recherche de grille
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}

# Dfinition du score  utiliser pour la recherche de grille
scoring = make_scorer(mean_absolute_error, greater_is_better=False)

# Dictionnaire pour stocker les scores moyens pour chaque horizon de prdiction
mean_scores_mae = []
mean_scores_corr = []

# Boucle sur chaque dlai pour effectuer la recherche de grille et obtenir les scores
delais = [delai_0days, delai_7days, delai_14days, delai_21days, delai_28days, delai_35days]
delai_names = ['0days', '7days', '14days', '21days', '28days', '35days']

for delai, name in zip(delais, delai_names):
    # Recherche de grille et validation croise en une tape
    grid_search = GridSearchCV(Reg_R, param_grid=param_grid, cv=ts_cv, n_jobs=n_jobs, scoring=scoring)
    grid_search.fit(delai, rain_Index)
    
    # Meilleur modle trouv par GridSearchCV pour ce dlai
    best_model = grid_search.best_estimator_
    
    # Obtenir les scores avec cross_val_score en utilisant le meilleur modle
    scores = cross_val_score(best_model, delai, rain_Index, cv=ts_cv, scoring=make_scorer(mean_absolute_error))
    # Calcul de la moyenne des scores pour ce dlai
    mean_score = np.mean(scores)
    mean_scores_mae.append(mean_score)
    
    scores_corr = cross_val_score(best_model, delai, rain_Index, cv=ts_cv, scoring=make_scorer(correlation_coefficient))
    mean_score_corr = np.mean(scores_corr)
    mean_scores_corr.append(mean_score_corr)
    
    #print(f'Best alpha for {name}:', grid_search.best_params_['alpha'])
    print(f'Mean MAE for {name}:', mean_score)
    print(f'Mean Correlation for {name}:', mean_score_corr)
# Sauvegarder les scores moyens dans un fichier .npy
S2S = np.array(mean_scores_mae)
np.save('/home/wester/Desktop/SST/donnees_models/mae_anom_rainfall_all_0-35days_Ridge.npy', S2S)
# Sauvegarder les coefficients de corrlation dans un fichier .npy
S2S_corr = np.array(mean_scores_corr)
np.save('/home/wester/Desktop/SST/donnees_models/corr_anom_rainfall_all_0-35days_Ridge.npy', S2S_corr)
```

# Multi-layer Perceptron Regressor (MLPRegressor)

**Le rseau de neurones artificiels (RNA)** est un modle de machine learning qui utilise des algorithmes d'apprentissage profond pour apprendre des relations entre les donnes. Il est compos de plusieurs couches de neurones, chacune d'entre elles tant connecte  la couche suivante. Chaque neurone reoit des entres pondres, les combine et passe ensuite cette somme pondre  travers une fonction d'activation pour produire une sortie.

Les paramtres du rseau de neurones sont des paramtres qui contrlent la structure et la faon dont le rseau de neurones apprend  partir des donnes.

```python
from sklearn.neural_network import MLPRegressor

# Cration d'un objet MLP Regressor
Reg_MLP = MLPRegressor(max_iter=2500)

# Dfinition des paramtres pour la recherche de grille
param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'alpha': [0.0001, 0.001, 0.01, 0.1]
}

# Utilisation du score MAE sans inversion de signe
scoring = make_scorer(mean_absolute_error)

# Dictionnaire pour stocker les meilleurs scores moyens pour chaque horizon de prdiction
mean_scores_mae = []
mean_scores_corr = []

# Boucle sur chaque dlai pour effectuer la recherche de grille et la validation croise
delais = [delai_0days, delai_7days, delai_14days, delai_21days, delai_28days, delai_35days]
delai_names = ['0days', '7days', '14days', '21days', '28days', '35days']

for delai, name in zip(delais, delai_names):
    # Effectuer une recherche de grille pour trouver les meilleurs hyperparamtres pour chaque dlai
    grid_search = GridSearchCV(Reg_MLP, param_grid=param_grid, cv=ts_cv, n_jobs=-1, scoring=scoring)
    grid_search.fit(delai, rain_Index)
    
    # Meilleur modle avec les hyperparamtres optimaux
    best_model = grid_search.best_estimator_
    
    # Validation croise avec le meilleur modle pour chaque dlai
    scores = cross_val_score(best_model, delai, rain_Index, cv=ts_cv, scoring=scoring)
    # Calcul de la moyenne des scores pour ce dlai
    mean_score = np.mean(scores)
    mean_scores_mae.append(mean_score)

    scores_corr = cross_val_score(best_model, delai, rain_Index, cv=ts_cv, scoring=make_scorer(correlation_coefficient))
    mean_score_corr = np.mean(scores_corr)
    mean_scores_corr.append(mean_score_corr)
    
    print(f'Mean MAE for {name} (best model):', mean_score)
    print(f'Mean Correlation for {name}:', mean_score_corr)

# Sauvegarder les scores moyens dans un fichier .npy
S2S = np.array(mean_scores_mae)
np.save('/home/wester/Desktop/SST/donnees_models/mae_anom_rainfall_all_0-35days_MLP.npy', S2S)
S2S_corr = np.array(mean_scores_corr)
np.save('/home/wester/Desktop/SST/donnees_models/corr_anom_rainfall_all_0-35days_MLP.npy', S2S_corr)
```
